# Car Market Data Analysis

This notebook explores the car listings scraped from mobile.de.

```python
import pandas as pd
import json

# Load scraped data
df = pd.read_json("cars_market_data.json")
print(f"Loaded {len(df)} cars")
df.head()
```

```python
# Available columns
df.columns.tolist()
```

```python
# Data types and missing values
df.info()
```

```python
# Quick statistics
df.describe()
```

## Data Cleaning

Convert price and mileage to numeric for analysis.

```python
# Clean price column (remove currency symbols and convert to numeric)
df['price_numeric'] = df['price'].str.replace(r'[^\d]', '', regex=True).astype(float)

# Clean mileage column
df['mileage_numeric'] = pd.to_numeric(df['mileage_km'], errors='coerce')

df[['title', 'price', 'price_numeric', 'mileage_km', 'mileage_numeric']].head(10)
```

## Price Analysis

```python
# Price distribution
df['price_numeric'].hist(bins=20)
print(f"Price range: {df['price_numeric'].min():,.0f} - {df['price_numeric'].max():,.0f}")
print(f"Average price: {df['price_numeric'].mean():,.0f}")
print(f"Median price: {df['price_numeric'].median():,.0f}")
```
